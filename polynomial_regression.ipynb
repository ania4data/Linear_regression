```python
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D   #3d plotting
import matplotlib.pyplot as plt
```

```python
X=[]
Y=[]

for line in open('data_poly.csv'):
    
    x1,y=line.split(',')
    x1=float(x1)
    X.append([(x1*x1),(x1),1])
    Y.append(float(y))

X=np.array(X)  
Y=np.array(Y)
```

```python
X
```

```python
#plt.scatter(X[:,0],Y)
plt.scatter(X[:,1],Y)
```

```python
w=np.linalg.solve(np.dot(X.T,X),np.dot(X.T,Y))
```

```python
w
```

```python
y_new=np.dot(X,(w.T))
```

```python
plt.scatter(X[:,1],Y)
plt.scatter(X[:,1],y_new)
```

```python
R_squared=1-(((Y-y_new).dot(Y-y_new))/((Y-Y.mean()).dot(Y-Y.mean())))
```
