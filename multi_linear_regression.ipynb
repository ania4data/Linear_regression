```python
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D   #3d plotting
import matplotlib.pyplot as plt
```

```python
X=[]
Y=[]

for line in open('data_2d.csv'):
    
    x1,x2,y=line.split(',')
    
    X.append([float(x1),float(x2),1])
    Y.append(float(y))

X=np.array(X)  
Y=np.array(Y)
```

```python
plt.scatter(X[:,0],Y)
plt.scatter(X[:,1],Y)
plt.scatter(X[:,2],Y)
```

```python
#or

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],Y)
```
