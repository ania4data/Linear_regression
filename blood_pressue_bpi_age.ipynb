```python

X=[]
Y=[]
i=0
for line in open('masctodos.csv'):
    
    #print(line.lstrip().split('\n')[0].split(';'))
    if (i>0):

        id,age,obese,bmi,wc,hc,whr,SBP,DBP,hyper=line.lstrip().split('\n')[0].split(';')
        if str(obese).upper()=='NO':
            obese_vec=0
        elif str(obese).upper()=='YES':
            obese_vec=1
            
        X.append([int(age),int(obese_vec),int(bmi),int(wc),int(hc),int(whr),1])
        Y.append([int(SBP),int(DBP)])

         
    i=i+1 
    #print(i)
X=np.array(X)  
Y=np.array(Y)    
```

```python
dim=np.shape(X)[0]
X_norm=np.zeros((dim,np.shape(X)[1]))
```

```python
for i in range(0,np.shape(X)[1]-1):

    X_norm[:,i]=(X[:,i]-np.mean(X[:,i]))/(np.max(X[:,i])-np.min(X[:,i]))

X_norm[:,np.shape(X)[1]-1]=1.0    
```

```python

def solve_linreg(X,Y,L2,lambda_):
    dim=np.shape(X)[1]
    if (L2==False):  
        w=np.linalg.solve(np.dot(X.T,X),np.dot(X.T,Y))
        return w
    elif (L2==True):
        w=np.linalg.solve(((np.eye(dim)*lambda_)+(np.dot(X.T,X))),np.dot(X.T,Y))
        return w
        
        
def y_new_cal(X,w):
    y_new=np.dot(X,(w.T))
    return y_new
    
 
```

```python
Data_=[]
Data_.append(['BP','reason','R_squared','RMSE','slope','intercept'])
for j in range(0,2):
    for i in range(0,6):
        w=solve_linreg(X[:,(i,np.shape(X)[1]-1)],Y[:,j],L2=False,lambda_=0)
        y_new=y_new_cal(X[:,(i,np.shape(X)[1]-1)],w) 
        R_squared=1-(((Y[:,j]-y_new).dot(Y[:,j]-y_new))/((Y[:,j]-Y[:,j].mean()).dot(Y[:,j]-Y[:,j].mean())))  
        delta=((Y[:,j]-y_new))
        N=np.shape(X)[0]
        rmse=np.sqrt((delta.dot(delta))/N)
        print('i, j, R_squarred, RMSE, w0,w1 \n',i,j,R_squared,rmse,w[0],w[1]) 
        print('0:Age;1:Obese;2:bmi;3:wc;4:hc;5:whr')
        Data_.append([j,i,R_squared,rmse,w[0],w[1]])
        fig=plt.figure()
        plt.scatter(X[:,i],Y[:,j])
        plt.plot(X[:,i],y_new)

        plt.xlabel('Reason')
        if j==0:
            plt.ylim(70,100)
            plt.ylabel('Low BP')
        if j==1:
            plt.ylim(70,200)
            plt.ylabel('High BP')
        plt.show()
        
```
